```python
import pandas as pd
import glob

# a = pd.read_csv('./data/coin_Bitcoin.csv')
# a

csv_files = glob.glob('./data/coin_*.csv')
dfs = [pd.read_csv(f) for f in csv_files]
all_coins = pd.concat(dfs, ignore_index=True)
all_coins = pd.DataFrame(all_coins)
all_coins[['High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']] = all_coins[['High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']].astype(float)
all_coins
```

```python
model = genai.GenerativeModel("gemini-1.5-flash",
                              system_instruction=f"""
                              Bạn tên là CryAI, một trợ lý AI có nhiệm vụ hỗ trợ giải đáp thông tin cho khách hàng về tiền ảo.
                              
                              Các chức năng mà bạn hỗ trợ gồm:
                              bạn có nguồn dữ liệu này {all_coins}
                              1. Gioi thiệu về tiền ảo dua vao file csv
                              2. hướng dẫn các loại tien ảo có lợi nhuận cao nhất
                              3. hướng dẫn các loại tiền ảo có rủi ro cao nhất
                              
                              Đối với các câu hỏi ngoài chức năng mà bạn hỗ trợ, trả lời bằng 'Tôi đang không hỗ trợ chức năng này. 
                              """)
```

```python
prompt = "high values of aave"
response = model.generate_content(prompt)
response.text
```
